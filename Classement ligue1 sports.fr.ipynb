```python
path_to_chromedriver = '/Users/Martin/Desktop/NSM/Paris/chromedriver' # change path as needed
browser = webdriver.Chrome(executable_path = path_to_chromedriver)
browser.implicitly_wait(30)
url = ['http://www.les-sports.info/football-championnat-d-angleterre-premier-league-2011-2012-resultats-eprd31128.html',
'http://www.les-sports.info/football-championnat-d-angleterre-premier-league-2012-2013-resultats-eprd37141.html',
'http://www.les-sports.info/football-championnat-d-angleterre-premier-league-2013-2014-resultats-eprd46863.html']
ref=['premierleague 2011/2012','premierleague 2012/2013','premierleague 2013/2014']
calendrier=collections.defaultdict(str)
effectuée=[]
for site in range(int(len(url))):
    browser.get(url[site])
    day=38
    ahh=browser.find_element_by_xpath('//*[@id="rencontres"]/table/tbody/tr[1]/td/h6').text
    calendrier[str(day)+'journée '+str(ref[site])]=ahh.replace('1er','1')
    effectuée.append(str(day)+'journée '+str(ref[site]))    
    for journée in range(37):
        time.sleep(3)
        browser.find_element_by_xpath('//*[@id="rencontres"]/div[1]/div[1]/div/a').click()
        day=37-journée
        time.sleep(3)
        ahh=browser.find_element_by_xpath('//*[@id="rencontres"]/table/tbody/tr[1]/td/h6').text
        calendrier[str(day)+'journée '+str(ref[site])]=ahh.replace('1er','1')
        effectuée.append(str(day)+'journée '+str(ref[site]))
        
        
    time.sleep(3)
    
#traitement de la date: je transforme 21 Avril 2012 en 2012,4,21 puis je retire un jour

moisdelan=['Janvier','Février','Mars','Avril','Mai','Juin','Juillet','Août','Septembre','Octobre','Novembre','Décembre']

for i in range(int(len(effectuée))):
    aza=calendrier[effectuée[i]].split(' ')
    mois2=moisdelan.index(aza[1])+1
    calendrier[effectuée[i]]= datetime(year=int(aza[2]),month=int(mois2),day=int(aza[0])) - timedelta(days=1) 

for i in range(int(len(effectuée))):
    print(calendrier[effectuée[i]])
    print(effectuée[i])
```
